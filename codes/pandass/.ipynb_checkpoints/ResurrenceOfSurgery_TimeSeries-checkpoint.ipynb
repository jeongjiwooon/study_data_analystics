```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rc('font',family='Malgun Gothic')
import seaborn as sns
import pandas as pd
import scipy.stats as stats
import numpy as np
```

```python
df_ROS=pd.read_csv('../../datasets/RecurrenceOfSurgery.csv')
df_ROS[:3]
```

- 접근방식
- 표현하는 데이터 타입
- 시각화 종류
    - 단일변수 / 연속형 : histplot
    - 단일변수 / 범주형 : barplot
    - 단일변수 / 순서형 : lineplot
- 현 데이터로 시각화 가능 여부
    - 한 chart에 동시 표현 여부 : 레코드 부리 되어 있어야 함.
    - 단일 변수 표현 시 주의점 : count 대치하는 변수 여부

```python
# 필요한 변수만 추출
df_ROS_extract = df_ROS[['환자ID','입원일자','퇴원일자']]
df_ROS_extract[:3]
```

```python
df_ROS_extract['입원일자_datatime']= pd.to_datetime(df_ROS_extract['입원일자'].astype(str))
df_ROS_extract['퇴원일자_datatime']= pd.to_datetime(df_ROS_extract['퇴원일자'].astype(str))
```

```python
df_ROS_extract.info()
```

### 주차별 환자 입원/퇴원 추이 시각화 및 인사이트
- 표현하는 데이터 타입 : 연속형, 단일변수
- 시각화 종류 : lineplot

### 요일별 환자 입원/퇴원 추이 시각화 및 인사이트
- 표현하는 데이터 타입 : 범주형, 단일변수
- 시각화 종류 : barplot
- 레코드 갯수로 count

```python
df_ROS_extract['입원일자_datatime_day'] = df_ROS_extract['입원일자_datatime'].dt.day_name()
df_ROS_extract['퇴원일자_datatime_day'] = df_ROS_extract['퇴원일자_datatime'].dt.day_name()
```
