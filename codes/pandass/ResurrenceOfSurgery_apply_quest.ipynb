```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rc('font',family='Malgun Gothic')
import seaborn as sns
```

```python
df_ROS = pd.read_csv('../../datasets/RecurrenceOfSurgery.csv')
df_ROS[:10]
```

```python
df_ROS.columns
```

```python
def miter (data) :
    return data/100
```

```python
df_ROS['신장_m']= df_ROS['신장'].apply(miter)
df_ROS['신장_m']
```

```python
df_ROS['BMI'] = df_ROS['체중'] / (df_ROS['신장_m'] * df_ROS['신장_m'])
```

```python
df_ROS['BMI']
```

```python
def BMI(data):
    if (data) < 18.5 :
        return '저체중'
    elif 18.5 <= (data) < 23 :
        return '정상체중'
    elif 23 <= (data) < 25 :
        return '과체중'
    elif 25 <= (data) < 30 :
        return '비만'
    else :
        return '고도비만'
```

```python
df_ROS['BMI_grade']=df_ROS['BMI'].apply(BMI)
```

```python
df_ROS[['환자ID','신장','체중','BMI','BMI_grade']]
```

```python
df_BMI_pivot=pd.pivot_table(data=df_ROS, index='BMI_grade', aggfunc=len)
df_BMI_pivot
```

```python
df_BMI_pivot['BMI_grade'] = df_BMI_pivot.index
```

```python
df_BMI_pivot[:10]
```

```python
sns.barplot(data=df_BMI_pivot, x='BMI', y='회원ID')
```
