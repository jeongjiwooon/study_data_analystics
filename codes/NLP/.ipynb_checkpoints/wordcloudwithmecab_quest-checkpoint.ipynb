```python
from wordcloud import WordCloud
```

```python
sentences = [
    "이 영화는 정말로 감동적이었습니다. 배우들의 연기력과 스토리텔링이 훌륭하며, 마지막 장면은 정말로 눈물을 자아내었습니다.",
    "시간 가는 줄 모르고 봤네요. 몰입도가 너무 높아서 이야기에 완전히 흡수되었습니다. 추천합니다!",
    "배경 음악부터 촬영 기법까지 모든 면에서 완벽한 영화입니다. 저는 이런 작품을 기다렸던 것 같아요.",
    "스릴 넘치는 액션과 긴장감 있는 플롯으로 지루할 틈 없이 재미있게 볼 수 있었습니다. 꼭 보세요!",
    "캐릭터들의 감정 변화와 내면 성장을 그린 이야기가 인상 깊었습니다. 여운이 오래 남았어요.",
    "대단한 시나리오와 함께 최고의 캐스팅! 배우들의 연기력은 정말 대단하다고 생각합니다.",
    "영상미와 비주얼 효과가 화려하고 아름다웠습니다. 시각적으로도 매우 만족스러웠어요.",
    "천재적인 감독의 솜씨가 돋보이는 작품입니다. 전체적인 분위기와 스토리 전개가 완벽하게 어우러져 있습니다.",
    "영화 속 캐릭터들의 대사와 유머 감각이 정말 좋았습니다. 웃음 가득한 시간을 보낼 수 있었습니다.",
    "심오한 주제를 다룬 영화로서 생각할 거리를 많이 주었습니다. 여운이 오래 남을 것 같아요."
]
```

```python
from mecab import MeCab
mecab = MeCab()
```

```python
mecab.nouns(sentences[0])
```

```python
morphs_list = list()
for sentence in sentences:
    morphs = mecab.nouns(sentence)
    morphs_list.append(morphs)
morphs_list
```

```python
morphs_list = list()
for sentence in sentences:
    morphs = mecab.nouns(sentence)
    morphs_list = morphs_list + morphs
morphs_list
```

```python
text = ' '.join(morphs_list)
text
```

```python
import re
pattern = r'[는로의과이은을가에한와와]'
morphs_list = re.sub(pattern=pattern, repl='', string=text) # params : 변경대상, 대상 대체, 어떤것을
morphs_list
```

```python
from collections import Counter
nouns_count = Counter(morphs_list)
nouns_count
```

```python
font_path = 'C:/Windows/Fonts/malgun.ttf'

wordCloud = WordCloud(font_path=font_path, background_color='white') #인스턴스화 
wordCloud.generate_from_frequencies(nouns_count)
```

```python
import matplotlib.pyplot as plt
plt.imshow(wordCloud)
plt.axis('off') 
plt.show()
```
