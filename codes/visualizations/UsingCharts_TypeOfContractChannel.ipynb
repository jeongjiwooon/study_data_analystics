### 범주형 : 주로 문자, 항목으로 구성

```python
df_TOCC['type_of_contract'].value_counts()
```

```python
sns.countplot(data=df_TOCC, x='type_of_contract')
plt.title('계약 종류')
plt.show()
```

```python
df_TOCC['type_of_contract2'].value_counts()
```

```python
sns.countplot(data=df_TOCC, x='type_of_contract2')
plt.title('계약 상품 종류')
plt.show()
```

```python
sns.countplot(data=df_TOCC, x='type_of_contract2', hue='product') #product로 group by 후 출력
plt.title('계약 상품 종류')
plt.show()
```

```python
sns.countplot(data=df_TOCC, y='type_of_contract2') #x 에서 y축으로 변경
plt.title('계약 상품 종류')
plt.show()
```

```python
df_TOCC['bank'].value_counts()
```

```python
sns.countplot(data=df_TOCC,y='bank')
plt.show()
```

```python
# 거래은행 중 상위 7위 확인하자. bankcolumn 항목 serise의 인덱스를 list 형태로 변환
order_list = df_TOCC['bank'].value_counts().index.to_list()[:7]
order_list
```

```python
# 필요 항목만 출력 가능 sns.countplot(data=df_TOCC,y='bank', order=['국민은행','롯데카드','BC카드'])
sns.countplot(data=df_TOCC,y='bank',order=order_list) 
plt.show()
```
